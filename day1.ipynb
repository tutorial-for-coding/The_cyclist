```python

#how to make sure the game doesn't exit automatically
import pygame

#constant all caps
SCREEN_WIDTH= 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 1")

#game loop
running = True

while running == True:
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()


```

```python

#how can we display the sun, but opps the background is not rendering
import pygame

SCREEN_WIDTH= 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 1")

running = True

sun_img = pygame.image.load("./images/sun.png")
background_img =  pygame.image.load("./images/sky.png")

while running == True:
    
    screen.blit(sun_img, (100, 100))
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()
    pygame.display.update()
```

```python

###now the background renders but where is the sun?
import pygame

SCREEN_WIDTH= 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 1")

running = True

sun_img = pygame.image.load("./images/sun.png")
background_img =  pygame.image.load("./images/sky.png")

while running == True:
    
    screen.blit(sun_img, (100, 100)) #sun is rendered first

    screen.blit(background_img, (0, 0)) # then background on top of that so the sun is hidden
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()
    pygame.display.update()
```

```python

###how objects are rendered from top to bottom, top one first and on top of that the second one
import pygame

SCREEN_WIDTH= 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 1")

running = True

sun_img = pygame.image.load("./images/sun.png")
background_img =  pygame.image.load("./images/sky.png")

while running == True:
    
    screen.blit(background_img, (0, 0)) # background first
    screen.blit(sun_img, (100, 100)) #sun on top of the background

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()
    pygame.display.update()
```
