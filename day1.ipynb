```python
## Day 1: Basic setup for the game

# Importing the Pygame library so we can use its tools
import pygame  

# ------------------------------
# CONSTANTS (always written in ALL CAPS)
# These define the size of the game window
# ------------------------------
SCREEN_WIDTH = 600    # width of the game window in pixels
SCREEN_HEIGHT = 600   # height of the game window in pixels

# ------------------------------
# CREATE GAME WINDOW
# ------------------------------
# "set_mode" creates the game screen using the width and height we defined
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Set a title for the window so it shows on the top bar
pygame.display.set_caption("Day 1 - My First Game Window")
```

```python
# Day 1: Keeping the game window open

import pygame

SCREEN_WIDTH = 600   
SCREEN_HEIGHT = 600  

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Day 1 - My First Game Window")

# ------------------------------
# GAME LOOP
# ------------------------------
# 'running' is True, so the game keeps running
running = True

while running:
    # Check all the events happening (like key presses, mouse clicks, quit, etc.)
    for event in pygame.event.get():
        # If the player clicks the close button (X), stop the game loop
        if event.type == pygame.QUIT:
            running = False

# When the loop ends, close the game safely
pygame.quit()
```

```python
## Day 1: How to display an object on the screen
## and why it doesn't work without some fixes!

import pygame

SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Day 1 - Display an Object")

# ------------------------------
# LOAD IMAGES
# ------------------------------
# Load the sun image, scale it to 300x400 pixels, and keep transparency
sun_img = pygame.transform.scale(
    pygame.image.load("./images/sun.png"), (300, 400)
).convert_alpha()

# Load a background image (sky)
background_img = pygame.image.load("./images/sky.png")

# ------------------------------
# GAME LOOP
# ------------------------------
running = True
while running:
    
    # --- DRAWING STARTS ---
    # First, draw the background so it covers the screen
    screen.blit(background_img, (0, 0))  

    # Then, draw the sun on top of the background at position (100, 100)
    screen.blit(sun_img, (100, 100))  

    # IMPORTANT: Update the display so changes actually appear!
    pygame.display.update()
    # --- DRAWING ENDS ---

    # Handle events (like quitting)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

# Quit the game safely
pygame.quit()
```

```python
### how objects are rendered from top to bottom,
### the first one is drawn first, and then the next one goes on top of it
import pygame

SCREEN_WIDTH = 600
SCREEN_HEIGHT = 600

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 1")

running = True

# Load the sun image
sun_img = pygame.image.load("./images/sun.png")
# Load the background image
background_img = pygame.image.load("./images/sky.png")

while running == True:
    
    # Step 1: Draw the background first (fills the whole screen)
    screen.blit(background_img, (0, 0))

    # Step 2: Then draw the sun on top of the background
    # Now both images are visible in the right order
    screen.blit(sun_img, (100, 100))

    # Handle quit event
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    # Update the display to show the drawings
    pygame.display.update()
```
