```python
#define the key bindings, and introduce the use of keyboard

import pygame

SCREEN_WIDTH= 1280
SCREEN_HEIGHT = 720

PLAYER_X_POSITION = 300
PLAYER_Y_POSITION = 500 

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 3")

running = True

player_img = pygame.transform.scale(pygame.image.load('./Assets/Character/Frame 1.png'), (100, 140))
player_rect = player_img.get_rect(center = (PLAYER_X_POSITION,PLAYER_Y_POSITION) )
background_img =  pygame.transform.scale(pygame.image.load("./Assets/BG/bg 01.png") , (SCREEN_WIDTH, SCREEN_HEIGHT)) 


clock = pygame.time.Clock()

jumping = False

while running == True:
    
    dt = clock.tick(60) # dt in milli seconds

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    keys_pressed = pygame.key.get_pressed()

    if keys_pressed[pygame.K_SPACE]:
        jumping = True


    # Move 50 pixels per second
    if jumping:
        PLAYER_Y_POSITION = PLAYER_Y_POSITION - 5
    else:
        PLAYER_X_POSITION += 5
    screen.blit(background_img, (0, 0)) 
    player_rect = player_img.get_rect(center = (PLAYER_X_POSITION,PLAYER_Y_POSITION) )
    screen.blit(player_img, player_rect)

    pygame.display.flip()

##conclusion: we determine which key was pressed, check if its space, and if so then move the player up
```

```python
#introduce gravity

import pygame

SCREEN_WIDTH= 1280
SCREEN_HEIGHT = 720

PLAYER_X_POSITION = 300
PLAYER_Y_POSITION = 500 

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 3")

running = True

player_img = pygame.transform.scale(pygame.image.load('./Assets/Character/Frame 1.png'), (100, 140))
jumping_img = pygame.transform.scale(pygame.image.load('./Assets/Jump/Jump_frame 6.png'), (100, 140))

player_rect = player_img.get_rect(center = (PLAYER_X_POSITION,PLAYER_Y_POSITION) )
background_img =  pygame.transform.scale(pygame.image.load("./Assets/BG/bg 01.png") , (SCREEN_WIDTH, SCREEN_HEIGHT)) 

clock = pygame.time.Clock()

jumping = False

y_velocity = 20
y_gravity = 1
jump_height = 20

while running == True:
    
    dt = clock.tick(60) 

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    keys_pressed = pygame.key.get_pressed()

    if keys_pressed[pygame.K_SPACE]:
        jumping = True

    screen.blit(background_img, (0, 0))  

    # Move 50 pixels per second
    if jumping:
        PLAYER_Y_POSITION = PLAYER_Y_POSITION - y_velocity
        y_velocity = y_velocity - y_gravity
        if y_velocity < -jump_height:
            jumping = False
            y_velocity = jump_height
        player_rect = player_img.get_rect(center = (PLAYER_X_POSITION,PLAYER_Y_POSITION) )
        screen.blit(jumping_img, player_rect) 
    else:
        PLAYER_X_POSITION += 5
        player_rect = player_img.get_rect(center = (PLAYER_X_POSITION,PLAYER_Y_POSITION) )
        screen.blit(player_img, player_rect)

    pygame.display.flip()


##conclusion: we determine which key was pressed, check if its space, and if so then move the player up
```
