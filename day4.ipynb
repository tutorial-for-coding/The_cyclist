```python
# Day 4: Add a static enemy

import pygame

SCREEN_WIDTH = 1280
SCREEN_HEIGHT = 720

PLAYER_X_POSITION = 300
PLAYER_Y_POSITION = 500 

# NEW PART: enemy starting position constants
ENEMY_X_POSITION = 600  
ENEMY_Y_POSITION = 550  

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 4")

running = True

# Load player images
player_img = pygame.transform.scale(
    pygame.image.load('./Assets/Character/Frame 1.png'), (100, 140)
)
jumping_img = pygame.transform.scale(
    pygame.image.load('./Assets/Jump/Jump_frame 6.png'), (100, 140)
)

# Create player rect
player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))

# NEW PART: load enemy image and create rect
enemy_img = pygame.transform.scale(
    pygame.image.load('./Assets/Obstacles/Obstacle 02.png'), (100, 140)
)
enemy_rect = enemy_img.get_rect(center=(ENEMY_X_POSITION, ENEMY_Y_POSITION))

# Load background
background_img = pygame.transform.scale(
    pygame.image.load("./Assets/BG/bg 01.png"), (SCREEN_WIDTH, SCREEN_HEIGHT)
)

# Clock for frame rate
clock = pygame.time.Clock()

# Jump variables
jumping = False
y_velocity = 20
y_gravity = 1
jump_height = 20

while running:

    dt = clock.tick(60)

    # Handle quit events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    # Detect key presses
    keys_pressed = pygame.key.get_pressed()
    if keys_pressed[pygame.K_SPACE]:
        jumping = True

    # Draw background
    screen.blit(background_img, (0, 0))

    # Player movement and jumping logic
    if jumping:
        PLAYER_Y_POSITION -= y_velocity
        y_velocity -= y_gravity
        if y_velocity < -jump_height:
            jumping = False
            y_velocity = jump_height
        player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))
        screen.blit(jumping_img, player_rect)
    else:
        PLAYER_X_POSITION += 5
        player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))
        screen.blit(player_img, player_rect)

    # NEW PART: display the static enemy on the screen
    screen.blit(enemy_img, enemy_rect)

    # Update display
    pygame.display.flip()

## Conclusion:
# - We introduced a static enemy by loading its image and creating a rect
# - Displaying the enemy is as simple as blitting it on the screen
# - The enemy doesn't move yet, it's just rendered in the foreground
```

```python
# Day 4: Add an enemy with movement

import pygame

SCREEN_WIDTH = 1280
SCREEN_HEIGHT = 720

# NEW PART: enemy starting position constants
ENEMY_X_POSITION = 600 
ENEMY_Y_POSITION = 550  

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 4")

running = True

# Player constants and images
PLAYER_X_POSITION = 300
PLAYER_Y_POSITION = 500 

player_img = pygame.transform.scale(
    pygame.image.load('./Assets/Character/Frame 1.png'), (100, 140)
)
jumping_img = pygame.transform.scale(
    pygame.image.load('./Assets/Jump/Jump_frame 6.png'), (100, 140)
)

player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))

# NEW PART: enemy image and rect
enemy_img = pygame.transform.scale(
    pygame.image.load('./Assets/Obstacles/Obstacle 02.png'), (100, 140)
)
enemy_rect = enemy_img.get_rect(center=(ENEMY_X_POSITION, ENEMY_Y_POSITION))

# Background image
background_img = pygame.transform.scale(
    pygame.image.load("./Assets/BG/bg 01.png"), (SCREEN_WIDTH, SCREEN_HEIGHT)
)

clock = pygame.time.Clock()

jumping = False
y_velocity = 20
y_gravity = 1
jump_height = 20

while running:

    dt = clock.tick(60) 

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    keys_pressed = pygame.key.get_pressed()
    if keys_pressed[pygame.K_SPACE]:
        jumping = True

    screen.blit(background_img, (0, 0))  

    if jumping:
        PLAYER_Y_POSITION -= y_velocity
        y_velocity -= y_gravity
        if y_velocity < -jump_height:
            jumping = False
            y_velocity = jump_height
        player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))
        screen.blit(jumping_img, player_rect)
    else:
        PLAYER_X_POSITION += 5
        player_rect = player_img.get_rect(center=(PLAYER_X_POSITION, PLAYER_Y_POSITION))
        screen.blit(player_img, player_rect)

    # NEW PART: move the enemy to the left each frame
    ENEMY_X_POSITION -= 5
    # NEW PART: update enemy rect to match new position
    enemy_rect = enemy_img.get_rect(center=(ENEMY_X_POSITION, ENEMY_Y_POSITION))
    # Draw enemy
    screen.blit(enemy_img, enemy_rect)

    # Update display
    pygame.display.flip()

## Conclusion:
# - We make the enemy move by decreasing its X position each frame
# - Update its rect so the enemy image stays aligned with its position
# - The enemy now moves from right to left across the screen
```

```python
# Day 4: Creating a Player class using OOP

import pygame

# NEW PART: Define a Player class that inherits from pygame.sprite.Sprite
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        
        # NEW PART: list to store normal sprites for animation
        self.sprites = []
        self.current_sprite = 0
        self.sprites.append(
            pygame.transform.scale(
                pygame.image.load('./Assets/Character/Frame 1.png').convert_alpha(), 
                (100, 150)
            )
        )

        # NEW PART: list to store jump sprites
        self.jump_sprites = []
        self.jump_sprites.append(
            pygame.transform.scale(
                pygame.image.load('./Assets/Jump/Jump_frame 1.png').convert_alpha(), 
                (100, 140)
            )
        )
        
        # Set initial image to the first jump sprite
        self.image = self.jump_sprites[0]

        # NEW PART: position and rect
        self.character_x_position = 300
        self.character_y_position = 500
        self.rect = self.image.get_rect(center=(self.character_x_position, self.character_y_position))

        # NEW PART: jump variables
        self.y_velocity = 15
        self.y_gravity = 0.5
        self.jump_height = 15
        self.jumping = False


SCREEN_WIDTH = 1280
SCREEN_HEIGHT = 720

screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("day 4")

# NEW PART: Initialize the player object after initializing the screen
player_object = Player()

# NEW PART: Use a sprite group to manage the player
player = pygame.sprite.GroupSingle()
player.add(player_object)

running = True

background_img = pygame.transform.scale(
    pygame.image.load("./Assets/BG/bg 01.png"), (SCREEN_WIDTH, SCREEN_HEIGHT)
)

clock = pygame.time.Clock()

jumping = False
y_velocity = 20
y_gravity = 1
jump_height = 20

while running:

    dt = clock.tick(60)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
            pygame.quit()

    keys_pressed = pygame.key.get_pressed()
    if keys_pressed[pygame.K_SPACE]:
        jumping = True

    screen.blit(background_img, (0, 0))

    player.draw(screen)
    player.update()

    # Update display
    pygame.display.flip()

## Conclusion:
# - The Player class encapsulates all attributes and behaviors of the player
# - Using sprite groups makes drawing and updating the player easier
# - This sets the stage for adding more players or enemies as objects
```
